* Filter Rows

* Sales greater than 500

```python
high_sales = df[df['Sales']>500]
print("\n=== High Sales Orders (Sales > 500) ===")
high_sales.head()
```

* Furniture category only

```python
furniture = df[df["Category"] == "Furniture"]
print("\n=== Furniture Category ===")
furniture.head(5)
```

```python
west_region = df[df["Region"] == "West"]
print("\n=== West Region Orders ===")
west_region.head(5)
```

* Add Month column

```python
df["Month"] = df["Order Date"].dt.to_period("M")
```

* Monthly sales

```python
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
print("\n=== Monthly Sales ===")
monthly_sales.head()
```

* Region wise sales

```python
region_sales = df.groupby("Region")['Sales'].sum().reset_index().sort_values(by="Sales", ascending=False)
print("\n=== Region Wise Sales ===")
region_sales.head()
```

* Category wise sales

```python
category_sales = df.groupby("Category")["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False)
print("\n=== Category Wise Sales ===")
category_sales.head()
```

* Top 10 products by sales

```python
top_products = df.groupby("Product Name")["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False).head(10)
print("\n=== Top 10 Products by Sales ===")
top_products
```

* Top 10 states by sales

```python
top_states = df.groupby("State")["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False).head(10)
print("\n=== Top 10 States by Sales ===")
top_states
```
